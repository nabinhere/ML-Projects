```python
from sklearn import datasets
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
list(iris.keys())
```

```python
iris['data'].shape
```

```python
x = iris['data'][:,3:] #peta width, row all column 3 and remaining
y = (iris['target'] == 2).astype(np.int)  #1 if iris-virginica else 0
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(x,y)
```

```python
x_new = np.linspace(0,3,1000).reshape(-1,1)
y_predicted = log_reg.predict_proba(x_new)
```

```python
print(y_predicted)
```

```python
y_predicted.shape
```

```python
print(y_predicted[0][0] + y_predicted[0][1] )
print(y_predicted[1][0] + y_predicted[1][1] )
print(y_predicted[2][0] + y_predicted[2][1] )
```

```python
plt.plot(x_new, y_predicted[:, 1], 'g-', label = "Iris-Virginica")
plt.plot(x_new, y_predicted[:, 0], 'b--', label = 'Not Iris-Virginica')
plt.xlabel('Petal Width, cm')
plt.ylabel('probability')
plt.show()
```
