Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn.datasets
from sklearn.metrics import accuracy_score
from sklearn.linear_model import  LogisticRegression
```

```python
#loading the data from breast cancer dataset
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
#print(breast_cancer_dataset)
```

```python
# loading the data to a pandas dataframe
breast_cancer = pd.DataFrame(data = breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
breast_cancer.head()
```

```python
# adding 'target' column to the dataframe
breast_cancer['label'] = breast_cancer_dataset.target
```

```python
# printing the last five rows
breast_cancer.tail()
```

```python
# checking the number of rows and columns
breast_cancer.shape
```

```python
# checking for the missing values
breast_cancer.isnull().sum()
```

```python
# finding the statistical measures
breast_cancer.describe()
```

```python
breast_cancer['label'].value_counts()
```

1 = benign &&
0 = malignant

```python
breast_cancer.groupby('label').mean()
```

Separating the features and targets

```python
X = breast_cancer.drop('label', axis = 1)
Y = breast_cancer['label'] 
```

```python
# print(X)
```

```python
# print(Y)
```

Train-Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state = 69)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training - Logistic Regression

```python
# creating an instance of logistic regression class or creating a model
logreg_model = LogisticRegression()
```

```python
logreg_model.fit(X_train, Y_train)
```

Model Evaluation



>Accuracy score



```python
# Accuracy score on training data
training_data_prediction = logreg_model.predict(X_train)
tarining_data_accuracy = accuracy_score(Y_train, training_data_prediction)
print('Accuracy on training data = ', tarining_data_accuracy)
```

```python
# Accuracy score on test data
test_data_prediction = logreg_model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, test_data_prediction)
print('Accuracy on test data = ', test_data_accuracy)
```

Building a predictive system


```python
input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# input data is a tuple, we need to convert it into a numpy array before further processing
input_data_as_numpy_array = np.asarray(input_data)

#reshaping the data as we are predicting for one sample only
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = logreg_model.predict(input_data_reshaped)
print(prediction)

if prediction[0] == 0:
  print('Malignant')
else:
  print('Benign')
```
